```python
import os
import glob
def modify_txt_files(directory_path):
    """
    Проходится по всем .txt файлам в указанной директории и заменяет
    в начале каждой строки указанные префиксы.

    Args:
        directory_path (str): Путь к директории с .txt файлами.
    """

    replacements = {
        'rca': '0',
        "pborca": "2",
        "pda": "1"
    }

    # Проверяем, существует ли директория
    if not os.path.isdir(directory_path):
        print(f"Ошибка: Директория '{directory_path}' не найдена.")
        return

    # Ищем все .txt файлы в директории (включая поддиректории, если нужно)
    # Для поиска только в текущей директории: os.path.join(directory_path, '*.txt')
    # Для поиска рекурсивно: os.path.join(directory_path, '**', '*.txt'), recursive=True
    txt_files = glob.glob(os.path.join(directory_path, '*.txt'))

    if not txt_files:
        print(f"В директории '{directory_path}' не найдено .txt файлов.")
        return

    print(f"Найдено {len(txt_files)} .txt файлов для обработки.")

    for filepath in txt_files:
        try:
            with open(filepath, 'r', encoding='utf-8') as file:
                lines = file.readlines()

            modified_lines = []
            changed_in_file = False
            for line in lines:
                original_line = line
                for old_prefix, new_prefix in replacements.items():
                    if line.startswith(old_prefix):
                        # Заменяем только префикс, остальную часть строки оставляем
                        line = new_prefix + line[len(old_prefix):]
                        changed_in_file = True
                        break # Переходим к следующей строке, так как замена для этой уже сделана
                modified_lines.append(line)

            if changed_in_file:
                with open(filepath, 'w', encoding='utf-8') as file:
                    file.writelines(modified_lines)
                print(f"Файл '{filepath}' успешно изменен.")
            else:
                print(f"В файле '{filepath}' не найдено строк для замены.")

        except Exception as e:
            print(f"Ошибка при обработке файла '{filepath}': {e}")

    print("\nОбработка завершена.")
```

```python
modify_txt_files('val/labels/')
```
